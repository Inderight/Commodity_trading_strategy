```python
ret.index = pd.to_datetime(ret.iloc[:,0])
ret.pop('Unnamed: 0')
ret
```

```python
df_nre =((new_nre['2019-06':'2021']*0.01)+1).cumprod()
```

```python
df_nre['new_strat'] =  (((((df_nre.iloc[:,0])*(df_nre.iloc[:,0]))*(df_nre.iloc[:,0])+ df_nre.iloc[:,1] + df_nre.iloc[:,2] + df_nre.iloc[:,3]\
                     +df_nre.iloc[:,4]+df_nre.iloc[:,5]+df_nre.iloc[:,6]+df_nre.iloc[:,7]+ df_nre.iloc[:,8]\
                     +df_nre.iloc[:,9])/(9+df_nre.iloc[:,0]*ret.iloc[:,0])))
df_nre
```

```python
nre_port= ((new_nre['2019-06':'2020']*0.01)+1).cumprod()['nre_port']
ret['NRE_Portfolio']=nre_port
ret['New_strategy'] =df_nre['new_strat']
ret.dropna(inplace = True)
ret
```
